```python
import pandas as pd
df = pd.read_csv ('data.csv')   #add you own data
print (df)
```

```python
df.plot()
```

```python
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
linearmodel = LinearRegression()
data= df[['X1','X2','X3','X4','X5']]

linearmodel.fit(data,df['Y'])
print(linearmodel.coef_)

preds = linearmodel.predict(data)
#print(preds - df['Y'])
#linearmodel.intercept_
linear= r2_score(df['Y'], preds)
print(linear)
```

```python
from sklearn import linear_model
ridgemodel = linear_model.RidgeCV()
ridgemodel.fit(data,df['Y'])
print(ridgemodel.coef_)
#ridgemodel.intercept_
ridgepred= ridgemodel.predict(data)
#print(ridgepred- df['Y'])
#ridgemodel.alpha_
ridge= r2_score(df['Y'], ridgepred)
print(ridge)
```

```python
from sklearn import linear_model
lasso = linear_model.LassoCV()
lasso.fit(data,df['Y'])
print(lasso.coef_)
#lasso.intercept_
preds3= lasso.predict(data)
#print(preds3- df['Y'])
lassor2=r2_score(df['Y'], preds3)
print(lassor2)
#lasso.alpha_
```

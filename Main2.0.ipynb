```python
import pandas as pd
import numpy as np
from scipy.interpolate import griddata
import matplotlib.pyplot as plt



x = df.iloc[:,0].values
y = df.iloc[:,1].values
Gene1 = df.iloc[:,2].values
Gene2 = df.iloc[:,3].values
Gene3 = df.iloc[:,4].values
Gene4 = df.iloc[:,5].values
Gene5 = df.iloc[:,6].values
Gene6 = df.iloc[:,7].values
Gene7 = df.iloc[:,8].values


#print(x)
#print(y)
#print(Gene1)
print(Gene2)
```

```python
def plot_contour(x,y,z,resolution,contour_method):
    resolution = str(resolution)+'j'
    X,Y = np.mgrid[min(x):max(x):complex(resolution),   min(y):max(y):complex(resolution)]
    points = [[a,b] for a,b in zip(x,y)] # efficeint iterating over data
    Z = griddata(points, z, (X, Y), method=contour_method)
    return X,Y,Z

X,Y,Z1 = plot_contour(x,y,Gene1,50,'linear')
X,Y,Z2 = plot_contour(x,y,Gene2,50,'linear')
X,Y,Z3 = plot_contour(x,y,Gene3,50,'linear')
X,Y,Z4 = plot_contour(x,y,Gene4,50,'linear')
X,Y,Z5 = plot_contour(x,y,Gene5,50,'linear')
X,Y,Z6 = plot_contour(x,y,Gene6,50,'linear')
X,Y,Z7 = plot_contour(x,y,Gene7,50,'linear')

with plt.style.context("default"):
    fig, a1 = plt.subplots(figsize=(13,8))
    fig, a2 = plt.subplots(figsize=(13,8))
    fig, a3 = plt.subplots(figsize=(13,8))
    fig, a4 = plt.subplots(figsize=(13,8))
    fig, a5 = plt.subplots(figsize=(13,8))
    fig, a6 = plt.subplots(figsize=(13,8))
    fig, a7 = plt.subplots(figsize=(13,8))
    
    a1.scatter(x,y, linewidth=1)
    a2.scatter(x,y, linewidth=1)
    a3.scatter(x,y, linewidth=1)
    a4.scatter(x,y, linewidth=1)
    a5.scatter(x,y, linewidth=1)
    a6.scatter(x,y, linewidth=1)
    a7.scatter(x,y, linewidth=1)
    
    a1.contourf(X,Y,Z1)
    a2.contourf(X,Y,Z2)
    a3.contourf(X,Y,Z3)
    a4.contourf(X,Y,Z4)
    a5.contourf(X,Y,Z5)
    a6.contourf(X,Y,Z6)
    a7.contourf(X,Y,Z7)
```
